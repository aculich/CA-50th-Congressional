### Importing Aggregated Data


```python
#specify datatypes so FIPS numbers, etc. read in properly
variable_types = {'Geo.id': 'str', 'GEO.id2': 'str'}

df = pd.read_csv('Exports\Race_Est_MOEs_2017.csv', delimiter = ',', header=[0], dtype=variable_types)
#specify dtype with variable_types we created above
```

```python
df=df.drop(['Unnamed: 0'], axis = 1) #Dropping a duplicative column
df.head(1)
```

##### Selecting columns

```python
# selecting columns
df = df[[ "GEO.id", 
                "GEO.id2", 
                "Total", 
                "Not_Latino_e_pct", 
                "White_e_pct", 
                "Black_e_pct", 
                "Native_e_pct", 
                "Asian_e_pct", 
                "API_e_pct", 
                "Other_e_pct", 
                "Two_plus_e_pct", 
                "Latino_e_pct", 
                "Other_agg_e_pct", 
                "Not_Latino_MOE_pct", 
                "White_MOE_pct", 
                "Black_MOE_pct", 
                "Native_MOE_pct", 
                "Asian_MOE_pct", 
                "API_MOE_pct", 
                "Other_MOE_pct", 
                "Two_plus_MOE_pct", 
                "Latino_MOE_pct", 
                "Other_Agg_MOE_pct" ]] 
```

##### Opening up lookup table to merge with the primary data frame. This will drop rows outside of the geography. 

```python
#specify datatypes so FIPS numbers, etc. read in properly
variable_types = {'GEOID': 'str'}

# opening up lookup table
df_lu = pd.read_csv('Exports\CA_Congress_Lookup_50th.csv', delimiter = ',', header=[0], dtype=variable_types)
```

```python
# dropping double index
df_lu=df_lu.drop(['FID'], axis = 1)
df_lu['CA_Con_Dist_50'] = 1
```

```python
df_lu.rename(columns={"GEOID" : "GEO.id2"}, inplace=True)
df_lu.head(1)
```

##### Merging datasets

```python
result = pd.merge(df, df_lu, on=['GEO.id2'])
result
```

```python
total  = result['Total'].sum()
total
```

```python
result['Congress_Density'] = result['Total'].div(942523)
result
```

```python
result.sort_values('Congress_Density', ascending=False).head(3)
```

##### Identiying top 10 Latinx Census tracts

```python
latinx = result[['GEO.id2', 'Latino_e_pct']]
latinx.sort_values(by=['Latino_e_pct'], ascending=False, inplace=True)
latinx = latinx.head(10)
latinx.head(2)
```

##### Identifying "other" top 10 census tracts

```python
other_race = result[['GEO.id2', 'Other_agg_e_pct']]
other_race.sort_values(by=['Other_agg_e_pct'], ascending=False, inplace=True)
other_race = other_race.head(10)
other_race.head(2)
```

##### Dropping columns /delete to exclude from the heatmap 

```python
results2 = result.drop(result.iloc[:, 13:45:1], axis = 1) 
results2 = results2.drop(result.iloc[:, 0:1:1], axis = 1)
results2 = results2.drop(result.iloc[:, 3:4:1], axis = 1)
results2
```

```python
results3 =results2
#.sort_values('Latino_e_pct', ascending=False, inplace=True)
results3
```

```python
format_dict ={'White_e_pct':'{:.1%}', 'Black_e_pct':'{:.1%}', 'Native_e_pct':'{:.1%}',
              'Asian_e_pct':'{:.1%}', 'API_e_pct':'{:.1%}', 'Other_e_pct':'{:.1%}', 
              'Two_plus_e_pct':'{:.1%}', 'Latino_e_pct':'{:.1%}', 'Other_agg_e_pct':'{:.1%}'}

results3 = results2.style.format(format_dict).hide_index()\
                .background_gradient(cmap='Blues', high=.99, low=.2,)
                
```
